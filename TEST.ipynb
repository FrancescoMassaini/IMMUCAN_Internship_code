```python
from tysserand import tysserand as ty
import pandas as pd
IMC_all=pd.read_csv("/home/francesco.massaini/Downloads/IMC_all.csv")
patient = "LUNG-NSCLC2-0641"
nodes = IMC_all.loc[IMC_all['patient'] == patient, ['Pos_X', 'Pos_Y', 'celltypes', 'patient']]
display(nodes)
nodes = nodes.rename(columns={'Pos_X': 'x', "Pos_Y":"y"})
coords = nodes[["x", "y"]].values
pairs = ty.build_delaunay(coords)
display(pairs)
# we want to avoid isolated cells, so we link them to their 3 closest neighbors
pairs = ty.link_solitaries(coords, pairs, method='delaunay', min_neighbors=3)
#ax = ty.plot_network(coords, pairs, labels=nodes['celltypes'])
# plt.title("sample {} file {}".format(sample_id, filename))
ax, fig = ty.plot_network(coords, pairs, labels=nodes['celltypes'])
```
